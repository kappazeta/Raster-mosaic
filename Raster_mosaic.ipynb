```python
# Creation of the raster mosaic 

#Import needed libraries

import glob
import os
from PIL import Image
from math import ceil, floor
import re
import numpy as np
```

```python
#Create a list of images from a given directory

image_list = []
for filename in glob.glob("Directory_with_images/*.png*.png"):
    im=Image.open(filename)
    image_list.append(im) #The images seem to be not in the correct order
```

```python
# Generate a list of PNG files and put it in order

images = glob.glob("Directory_with_images/*.png*.png")
images.sort(key=lambda var:[int(x) if x.isdigit() else x for x in re.findall(r'[^0-9]|[0-9]+', var)])
print(images) #Everything seems to be in order
```

```python
# Create a function which rotates each image in the directory

def rotateImages(rotationAmt):
  # for each image in the current directory
  for image in glob.glob("Directory_with_images/*.png"):
    # open the image
    img = Image.open(image)
    # rotate and save the image with the same filename
    img.rotate(rotationAmt).save(image)
    # close the image
    img.close()
    
rotateImages(270) #In this case the images seem to be rotated 270Ëš counter clockwise
```

```python
# Set image parameters

frame_width = 10980
images_per_row = 22
padding=0


img_width, img_height = Image.open(images[0]).size
sf = (frame_width-(images_per_row-1)*padding)/(images_per_row*img_width)  
scaled_img_width = ceil(img_width*sf)               
scaled_img_height = ceil(img_height*sf)

number_of_rows = ceil(len(images)/images_per_row)
frame_height = ceil(sf*img_height*number_of_rows) 

new_im = Image.new('RGB', (frame_width, frame_height))

i,j=0,0
for num, im in enumerate(images):
    if num%images_per_row==0:
        i=0
    im = Image.open(im)
    #Here I resize the opened image, so it is no bigger than 100,100
    im.thumbnail((scaled_img_width,scaled_img_height))
    #Iterate through a 4 by 4 grid with 100 spacing, to place my image
    y_cord = (j//images_per_row)*scaled_img_height
    new_im.paste(im, (i,y_cord))
    print(i, y_cord)
    i=(i+scaled_img_width)
    j+=1

# Open the generated image
new_im.show()

# Define a directory where to save a new file, resolution, etc.
new_im.save("Directory_and_final_file_name", "PNG", quality=10980, optimize=True, progressive=True)
```
